# Abstract

The objective of our project was to analyze any observable relationships between certain music genres and the composition of MIDI files of songs belonging to the genre. A MIDI file is an electronic audio file which provides lossless audio encoding. Instead of using soundwaves to represent audio, MIDI files track information about individual notes played within a song, and includes attributes such as pitch, volume, speed, and duration, allowing for the  precise recreation of each song. We hypothesize that MIDI files for songs belonging to different genres will experience patterns that will be indicative of their genre and allow classification. 

We construct a dataset containing 11 total genres by combining feature information extracted from the Lakh Midi Dataset, the Million Song Dataset, and the MusicBrainz database. We analyze our dataset by passing in MIDI features such as song length, note pitch, and timesignature into three models: Logistic Regression (LR), Support Vector Machine (SVM), and K-Nearest Neighbors (KNN). We define our baseline to be a probability function $f(x) = genre$, which will take in a datapoint $x$ and randomly predict one of the 11 possible genres, with the probability of each genre prediction given with respect to the proportion the genre comprises in the dataset. The baseline yielded a test accuracy of 0.272. The models yielded significantly higher test accuracies of 0.395 for our LR model, 0.512 for SVM, and 0.474 for our KNN model. Through our model results, we conclude that while there exists a definitive relationship between MIDI features and song genre that allow our models to outperform the baseline, the relationships captured between features in our dataset may not have been sufficiently genre-separable to yield reliable predictive results.

## Background and Prior Work

The Musical Instrumental Digital Interface (MIDI) is a lightweight electronic audio transmission protocol that has been held as the industry standard for encoding and storing digital music soon after its introduction in 1983. Instead of encoding audio information through waveforms, as is the case for traditional MP3 file formats, MIDI files encode audio by logging information about each note played within a musical piece, including factors such as its pitch, volume, the speed at which it is played, and the duration that the note lasts for. Omitting instrumentals from its encoding schema allows MIDI files to store extremely precise musical information that is universally interpretable among digital music systems, leaving the implementation of actual instrumental sound up to the individual system. We seek to take advantage of MIDI’s precise representation format as a means to analyze different genres of music with fine granularity, and to search for any common structural patterns among songs of the same genre.

There have been several prior instances of genre classification through MIDI analysis<a href="cite_ref-1"></a>[<sup>1,</sup>](#cite_note-1)<a href="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). In Sean Hobin’s genre predictor, he analyzed classical, jazz, ragtime, and folk MIDI files and classified features, such as  present in each file<a href="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). It was observed that different features had different weights depending on the genre. With this information, Hobins was able to accurately predict song genres with up to 90% accuracy<a href="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). However, Hobins notes that there is a lot of room for improvement in his model<a href="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1), as the algorithms he used for feature extraction were incomplete and often buggy, and the classifications were constrained within a very limited range of genres. In another example of music classification with MIDI, Chet Gengy classifies the playing style of instruments using MIDI files by taking into account statistics such as the mean and standard deviation of pitch, volume, and duration of notes. However, Gengy’s model ran into issues with feature separability during feature extraction, and the accuracy of the model suffered from overlapping features, such as rhythm and acousticness[<sup>2</sup>](#cite_note-2).

Aside from genre classification, other relevant works apply the same concept of MIDI feature extraction to perform classifications such as composer identification<a href="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) or music emotion recognition<a href="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). One work extracted information on frame rolls, onset rolls, and velocity rolls from a piano-based MIDI dataset to identify distinct composing styles from different composers, the results of which were passed to a neural network for classification<a href="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Another work edited MIDI files to isolate specific components of a song, such as the chorus or melody, before performing feature extraction on pitch and rhythmic variability in conjunction with natural language processing on the accompanying song lyrics in order to output a measure of the musical emotion associated with the song<a href="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4).

Moving forward for our project, we plan to take the lessons learned from the prior work and improve upon their methodologies. The common issue throughout all of the prior work examined is the feature extraction step, and the features examined and the extraction methods used were observed to have a large impact on the performances of the resulting models. Due to the nature of MIDI encoded files, there are a large variety of approaches that we could choose to take to perform extraction. We will have to experiment with multiple methodologies and critically analyze our experimental results in order to ensure that we do not repeat the pitfalls observed from prior works.

# Data Cleaning/Processing

Although our main dataset contains most of the information that we need (with the exception of genre data), the information is fairly spread out. We had two corresponding datasets with a one to one mapping with one another, where one dataset contained the actual MIDI files, while the other contained metadata information (such as song title and artist) in an h5 file. A fair amount of processing was required to clean up and format the data, and our data extraction began with genre extraction and matching those genres to the correct songs and MIDI file paths.

Our primary dataset is the Lakh MIDI dataset, which contains MIDI files, but not genre information. We wrote a script that will extract the metadata for the artist and title of each song from the file, and send a query containing the information to the MusicBrainz API. Since MusicBrainz is an open-source database, genre labels for each song entry stores a list of user-submitted genres, as well as the number of user-submitted votes for that song. We count the number of votes for each genre associated with the song that we query, and return the top 3 most voted genres. This result is stored in a JSON file, where each entry contains the information for the artist, song title, top 3 genres, and the filename of the song.

With the JSON file containing genres, we now had to match each song and genre to a midi file. Initially, all the midi files we downloaded and extracted were in deeply nested files, and many of the songs had multiple midi files. So, we had to iterate through the nested files and pick the “best” midi file to represent a song, which we decided using a dataset provided with the Lakh MIDI dataset containing information about how well a midi file matched a song(confidence). For each song we chose the highest confidence midi file and was able to reduce the number of files we had to process from ~45,000 to ~30,000. We then added the midi file location to the previously created JSON file, and this allowed us to parse through midi files for unique songs.

To extract the data from the midi files, we had to use packages such as Mido and pretty_midi. From the midi files we extracted the following features:
- ‘song_len’ (float): Length of the song in seconds
- ‘num_instruments’ (int): The number of instruments
- ‘instruments’ (list): List of tuples containing instrument program numbers and whether or not they are drums
- ‘key_changes’ (dict): A dictionary of times and key numbers
- ‘time_signature_changes’ (dict): A dictionary of times and key signatures as strings
- ‘lyrics_len’ (int): The number of words in the song lyrics
- ‘tempo_changes’ (dict): A dictionary of times and tempos
- ‘tempo’ (float): The best estimate of the song’s overall tempo
- ‘beat_start’ (float): The time of the first beat
- ‘num_notes’ (int): The total number of notes in the song across all instruments
- ‘pitch_class_histogram’ (list): A histogram of the pitch classes(octaves) in the song

While parsing through the files to extract data, it seemed that a small portion (2-3%) of the midi files were corrupted/unusable, and with such a large dataset, we decided the best thing to do was to just omit them from our final dataset.

```python
""" 
Note the data output of this file is already provided in data/match_genres.json

This file will load genre data
     It creates a json file of data where each entry is a song
     The data has the artist name, song name, top 3 genres, and metadata filename for each song
     
Credits - Kai Lee
"""

import requests as re
from collections import defaultdict
import os
import tables
import json


"""
The function get_genre below loads the top n genres of a song given its artist name and song title
Test use is shown right after (commented out)
"""

# possible genres: https://musicbrainz.org/genres 
def get_genre(artist, song, top_genre_count=1):  
    """ 
    get_genre makes a MB query and return N most common genres for the song, N=top_genre_count 
    artist: string
    song: string 
    top_genre_count: the number of genres to return, in descending order from most to least common
    """
    mb_url = "https://musicbrainz.org/ws/2/release-group"
    
    params = {
        'query': f'artist:"{artist}" AND recording:"{song}""?inc=genres',
        'fmt': 'json',
    }

    response = re.get(mb_url, params=params)    # GET request

    if response.status_code != 200: 
        print(f"Error for query <{artist}, {song}>: {response.status_code}")
        return None
    
    data = response.json()
    genres = defaultdict(int)   # key: genre, value: number of votes (# labels applied) for that genre

    if not data['release-groups']:  # assert not empty
        print(f"No valid release groups for <{artist}, {song}> found")
        return None 

    for i in range(len(data['release-groups'])):    # scum
        if 'tags' in data['release-groups'][i] and data['release-groups'][i]['tags']:
            for tag in data['release-groups'][i]['tags']:
                genres[tag['name']] += tag['count']

    if not genres:  # assert not empty
        print(f"No valid tags for <{artist}, {song}> found, skipping")
        return None 

    # get N most upvoted genres
    sorted_genres = sorted(genres, key=genres.get, reverse=True)[:top_genre_count]  
    
    return sorted_genres


# test_release = 'cold weather'
# test_artist = 'glass beach'

# # returns 10 most upvoted genre labels for cold weather by glass beach
# print(get_genre(test_artist, test_release, 10)) 

"""
The function match_genres below loads the before mentioned data, and writes it to a .json file called match_genre.json stored in the data folder
"""


def match_genres(directory, output_fp, backup='../data/backup.json', num_genres=3):
    """
    match_genres matches genres to each h5 file 
    directory: root directory containing all nested h5 files
    output_fp: filepath of output json file containing genre matching data
    backup: filepath of backup json file for intermediate writes (uncomment code under intermediary save to use)
    num_genres: number of genres to match to each song
    """
    json_data = []  # list of dicts that will be stored as json strings
    for root, dirs, files in os.walk(directory):   # recurse through directory
        backup_data = []
        for file in files:
            hdf5_path = os.path.join(root, file)
            data = tables.open_file(hdf5_path, mode='r')
            try:
                artist = data.root.metadata.songs.cols.artist_name[:][0].decode('utf-8')
                song = data.root.metadata.songs.cols.title[:][0].decode('utf-8')
                genres = get_genre(artist, song, num_genres)
                if genres:
                    match = { 'artist': artist, 'song': song, 'genres': genres, 'filename': file}
                    json_data.append(match)
                    backup_data.append(match)
                
                print(f"artist: {artist}\tsong: {song}\tgenres: {genres}\tfilename: {file}")
            except Exception as e:  # generic catch-all, stops code from terminating halfway
                print(f"Error parsing info for file {hdf5_path}: {e}")
                data.close()

            data.close()

        # intermediary save - data is uglier, but prevents total loss of progress if runtime error is encountered
        if backup_data:
            with open(backup, 'a') as file:
                json.dump(backup_data, file, indent=4)

    with open(output_fp, 'w') as file:
        json.dump(json_data, file, indent=4)
    
    print('finished')
    
# Example use for this project
# match_genres('../data/lmd_matched_h5/', '../data/match_genre.json')
```

```python
# Credits - Garvey Li, Kate Vu
# Feature extraction method for a single midi file
def feature(fp):
    """
    Extracts features from a PrettyMidi object created from a midi file

    Parameters:
    fp (str): Midi filepath

    Output:
    sd (dict): Song dictionary containing midi features

    """

    sd = {}
    spm = pretty_midi.PrettyMIDI(fp)
    sd['song_len'] = float(spm.get_end_time())

    # MIDI has 127 program numbers
    sd['num_instruments'] = len(spm.instruments)
    # List of instruments as tuple: (program_number, is_drum)
    sd['instruments'] = [(str(i.program), i.is_drum)
                          for i in spm.instruments]
    
    

    sd['key_changes'] = {str(k.time): float(k.key_number)
                          for k in spm.key_signature_changes}
    
    # Formats time signature changes as a dictionary with key: tick, value: time signature
    sd['time_signature_changes'] = {float(ts.time): str(ts.numerator) + "/" + str(ts.denominator) 
                                     for ts in spm.time_signature_changes}
    
    sd['lyrics_len'] = len(spm.lyrics)
    
     
    # First array is when the tempo changes, second array is what it changes to 
    tc = spm.get_tempo_changes()
    # Dictionary of tempo changes and when they occur {time: tempo}
    sd['tempo_changes'] = dict(zip(tc[0], tc[1]))

    # Best tempo estimate from estimate_tempi(). FLoat
    sd['tempo'] = float(spm.estimate_tempo())

    # All beats in the song
    # sd['beats'] = spm.get_beats().tolist()
    
    # Time of the first beat in the song
    sd['beat_start'] = spm.estimate_beat_start()

    # Downbeat location in seconds
    #sd['downbeats'] = spm.get_downbeats().tolist()

    # Total number of notes played
    sd['num_notes'] = sum([len(i.notes) for i in spm.instruments])
    #len(spm.get_onsets())
    # Histogram of pitches in the song
    pitch_hist = spm.get_pitch_class_histogram().tolist()
    sd['pitch_class_histogram'] = pitch_hist

    
    # Periodic form of the song. Can use FFT
    # sd['synthesis'] = spm.synthesize().tolist()
    return sd
```

```python
# Credits - Garvey Li
# Feature extraction for midi files contained in a JSON file
def get_features(loaded_data, output_fp, backup_name, backup_intervals=1500):
    """
    Extracts features of midi files present in a JSON containing midi filepaths
    and writes the features to multiple JSON files

    Parameters: 
    loaded_data (list): List of dictionaries that contain a midi filepath
    output_fp (str): Output JSON filepath for the entire dataset. 
    backup_name (str): Name of backup JSON files. Do not include '.json' in name.
    backup_intervals (int): Dump data that has been extracted into a JSON file
        after every backup_intervals number of files parsed (Dump every
        1500 files parsed by default)

    """
    file_count = 0
    file_count_num = 0
    total_files_parsed = 0
    json_data = []  # list of dicts that will be stored as json strings
    for s in loaded_data:
        # Dump data into JSON file every 1500 parsed files by default
        if file_count == backup_intervals:
            print("Creating backup")
            print('Valid Data Size: ', len(json_data))
            with open(backup_name + str(file_count_num) + '.json', 'w') as file:
                json.dump(json_data, file, indent=4)
            file_count = 0
            file_count_num += 1
        # Some files will throw exceptions when extracting data due to 
        # corruption
        try:
            final_sd = s
            final_sd.update(feature(final_sd['midi_fp']))
            json_data.append(final_sd)
            file_count += 1
        except Exception as e:
            print("Corrupt file: ", final_sd['midi_fp'])
            print("Exception: ", e)
            print("Valid Data Size: ", len(json_data))
            print("Total Files Parsed: ", total_files_parsed)
        total_files_parsed += 1
    with open(output_fp, 'w') as file:
        json.dump(json_data, file, indent=4)
    print("Finished. Final valid data size: ", len(json_data))
    return json_data
```

```python
# Full dataset
df_raw = pd.read_json(open('data/midi_full.json'), orient='records')

# Adding additional features

# Adding overall time and key signatures
time_signatures = pd.read_json(open('data/time_sigs.json'), orient='records')
key_signatures = pd.read_json(open('data/key_sigs.json'), orient='records')
key_signatures['key_signature'] = key_signatures['key_signature'].astype(str) 

# Converting pitch class histogram list into a dataframe where each pitch
# is a column
pitch_columns = ['p' + str(i) for i in range(1, 13)]
pitch_hist_df_raw = pd.DataFrame(df_raw["pitch_class_histogram"].to_list(), \
                             index=df_raw.index, columns=pitch_columns)
df_raw.drop(columns=['midi_fp', 'metadata', 'pitch_class_histogram'], inplace=True)

# Number of drum instruments
df_raw['num_drums'] = df_raw['instruments'].apply(lambda x: sum([bool(d[1]) for d in x]))

# Adding key and time signature
df_raw = df_raw.merge(time_signatures, how='left', left_on='song_id', right_on='song_id')
df_raw = df_raw.merge(key_signatures, how='left', left_on='song_id', right_on='song_id')
# Adding pitch histograms
df_raw = df_raw.merge(pitch_hist_df_raw, how='inner', left_index=True, right_index=True)

# Separating instrument program number and bool value for one hot encoding
df_raw['temp_instruments'] = df_raw['instruments'].apply(lambda x: [i[0] for i in x])

# One hot encoding the instruments/program numbers
mlb = MultiLabelBinarizer(sparse_output=True)
df_raw = df_raw.join(
            pd.DataFrame.sparse.from_spmatrix(
                mlb.fit_transform(df_raw.pop('temp_instruments')),
                index=df_raw.index,
                columns=mlb.classes_))
one_hot_programs = df_raw.dtypes[df_raw.dtypes.values == df_raw.dtypes.loc['99']].index
df_raw[one_hot_programs] = df_raw[one_hot_programs].astype(int)
df_raw.rename(columns={'tempo': 'overall_tempo'}, inplace=True)

# One Hot Encode Time signatures
df_raw = pd.get_dummies(df_raw, columns=['time_signature'], prefix='time_sig')
df_raw .iloc[:, -6:] = df_raw .iloc[:, -6:].astype(int) # Convert to 0/1 instead of False/True

# One Hot Encode key signatures
df_raw = pd.get_dummies(df_raw, columns=['key_signature'], prefix='key_sig')
df_raw.iloc[:, -12:] = df_raw.iloc[:, -12:].astype(int) # Convert to 0/1 instead of False/True


print(df_raw.shape)
df_raw.head()
```

### Pitch Class

Another variable that may have a high correlation with genre is pitch class. 
Each pitch class (from P1-P12) represents a note in the 12 chromatic pitches,
and the chromatic pitches consist of the following notes 
('#' denotes sharp, 'b' denotes flat):

A, A#/Bb, B, C, C#/Db, D, D#/Eb, E, F, F#/Gb, G, G#/Ab 

With each genre, there might be various frequencies of each note that would
allow us to distinguish genres by the statistics of pitch frequencies. This 
may occur due to disproportionate usage of key signatures in different genres,
as each key signature only contains a portion of all notes in the 12 chromatic
pitches.

Now pitch class isn't actually OHE, but we can take the means for each one groupd by genre to achieve a similar effect.

```python
# The mean frequencies of each pitch class across genres

# Mean frequencies of pitch classes of each 'top_genre'
mean_pitches = df.groupby(by=['generalized_genre'])[pitch_columns].mean()
# Overall mean frequencies of pitch classes
overall_mean_pitches = pd.DataFrame([df[pitch_columns].mean()])
overall_mean_pitches['generalized_genre'] = 'All'
overall_mean_pitches.set_index('generalized_genre', inplace=True)
# Combining overall and genre means into one dataframe
mean_pitches = pd.concat([overall_mean_pitches, mean_pitches])
# Getting list of genre names
genres = mean_pitches.index.tolist()

fig_pmean, ax_pmean = plt.subplots(4, 3, figsize=(12, 8))

graph_count = 0
for i in range(4):
    for j in range(3):
        ax_pmean[i, j].bar(pitch_columns, mean_pitches.loc[genres[graph_count]])
        ax_pmean[i, j].set_ylim(0, 0.15)
        ax_pmean[i, j].set_title(genres[graph_count] + ' Means')
        ax_pmean[i, j].set_xlabel('Pitch Class/Note')
        ax_pmean[i, j].set_ylabel('Frequency')
        graph_count += 1

plt.subplots_adjust(hspace=1)
```

From the mean distributions of notes, it is apparent that there is a general
trend in which pitch classes 1, 3, 5, 8, 10, and 12 generally have a high frequency.
The reason for this may lie in chord structures. A chord consisting of
pitch classes 1, 3, and 5 may be a major chord, as well as a chord consisting
of pitch classes 8, 10, and 12. Genres such as `Classical`, `Hip-Hop/Rap`, and
`R&B/Soul` have a slightly more uniform distribution compared to the other genres,
so looking at pitch class alone may allow us to distinguish those 3 genres from the rest.
Note that there may be other distinguishing features in each
histogram that are hard to visually identify.

## Data Modeling and Analysis

To address our hypothesis, we decided to determine if there was a relationship between the MIDI data and the genre of a song by training and evaluating multiclass classification models such as multi-class logistic regression, KNN, and SVC. We will evaluate these models using accuracy and weighted F-1 score since there is a severe class imbalance.

#### Train, Validation, Test Splits

Before we create our models, we have to create training, validation, and testing splits in order to evaluate our models. Our training set contains 80% of the data, and the validation and test sets contain 10% of the data, each. For each split, we have to stratify the splits in order to maintain the distributions of `generalized_genre` within each set due to the high class imbalance in our data (Subsampling and oversampling would not work in this case because the differences in genre counts is too extreme).

```python
# Imports
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegressionCV, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

import warnings
warnings.filterwarnings('ignore')
```

However, copyright issues may not be a problem since we are not playing the actual songs, but it's essential to be cautious about how we use copyrighted data. We will ensure that our use falls within fair use principles and does not infringe on the rights of copyright holders. If possible, we will use openly licensed MIDI files or data from sources that grant permission for research use. We would need to explore the distribution of songs to see if there are any biases. There are a lot of potential biases with songs. For example, what if the data has songs mainly from male artists or what if the majority of the songs are pop. The genres could be unevenly distributed. In our EDA, we can analyze the distributions of these potential bias terms. We will analyze the components of each data point (song), such as the artist characteristics, genres, etc. In terms of copyright issues, this is not a problem as our analysis never needs to play the actual songs. Moreover, the dataset we use is already publicly available and free to download.

# Discussion and Conclusion

We found that there are common structural patterns within songs of the same genre, revealing an observable relationship between music genres and the composition of MIDI files. We implemented a feature extraction method for MIDI files, capturing various aspects such as song length, number of instruments, key changes, time signature changes, lyrics length, tempo changes, and pitch class. The extracted features were utilized to create a dataset for further analysis. Outliers were removed to enhance the accuracy of predictive models for our dataset. The distribution of genres in the dataset was observed to be heavily skewed, with rock, electronic/dance, and pop genres dominating. 

These features’ distributions revealed some differences between genres, showing possible predictors of genre classification. In addition, we identified an effect of pitch classes and time signatures on genre differentiation. For example, classical music shows specific characteristics in time signatures, and pitch class distributions demonstrate genre-specific trends. Overall, features identified as relevant can be used to build a predictive model for music genre classification. We then constructed three different models based on our data: KNN, LogReg, and SVM. All three models showed an accuracy close to 50%, with the SVM model performing the best, meaning the models can predict genre but not efficiently. We think there are multiple factors that affect this, such as the features we used for data extraction and the datasets being skewed towards certain genres. 

With the size of the dataset came many limitations on time. It took upwards of 30 hours to extract the fairly basic features for ~40,000 MIDI files. There were some more complex features such as the wave synthesis of the song that we could've extracted and utilized, but extracting that feature alone took almost 10 times as long. With the save synthesis of each song, we could've identified common patterns in waves within each genre, which is a much more comprehensive feature than features such as key signature and tempo. However, even with the features we were able to extract, we were able to identify that there was some relationship that exists between the features and the genre.
